```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

d = {} 
numero=5080
#dataframe=None
datosrfinal = []
for count in range(0,10):
    URL='https://repository.usc.edu.co/handle/20.500.12421/'+str(numero)+'?show=full'
    estructura = requests.get(URL,verify=False).text
    contenido = BeautifulSoup(estructura, "lxml")
    #Lista para capturar los tr del documento html
    dastosr=[]

    for datos in contenido.find_all('tr'):
        validador=datos.text.replace('\n', '').strip()
        if validador.startswith('dc.contributor.advisor'):
            advisor=validador[22:]
        if validador.startswith('dc.subject.keyword'):
            dastosr.append(validador[18:])
        if validador.startswith('dc.subject'):
            l=validador[10:].replace('.keyword', '').strip()
            l=l.replace('es_ES', '').strip()
            dastosr.append(l)
        #se quitan duplicados de la lista
        for i in dastosr:
            if i not in datosrfinal:
                datosrfinal.append(i)
        #construcción de un diccionario para un dataframe del resultado final
        for i in datosrfinal:
            d[i] = advisor
numero+=1
dataframe=pd.Series(d).to_frame('Advisor').reset_index()
dataframe=dataframe.rename(columns={'index':'Keyword'})
dataframe = dataframe[["Advisor",'Keyword']]
```
