```python
#Primera parte, validada, ok
#DEFINICIÓN DE ÍNDICES
#Indice i
alimento=['AZUCAR','ARROZ','ACEITE','ATUN']
suplemento=['PANELA']
#Indice j
kits=[1,2,3]

#VARIABLES DE DECISIÓN
#Cantidad de alimento i asignado al kit j
X = LpVariable.dicts("Cantidad",[(i,j)for i in alimento for j in kits],lowBound=0)
k = LpVariable.dicts("Kit",[(j)for j in kits],lowBound=0,cat='Integer')
HINV=LpVariable.dicts("Holgura_Inventario",[(i)for i in alimento],lowBound=0,cat='Integer')

#PARÁMETROS
#Cantidad de alimento empleado i en conformar el kit tipo j
C={1:{'AZUCAR':1,'ARROZ':2,'ACEITE':0,'ATUN':1},
  2:{'AZUCAR':2,'ARROZ':0,'ACEITE':1,'ATUN':1},
  3:{'AZUCAR':1,'ARROZ':1,'ACEITE':0,'ATUN':1}}

INV={'AZUCAR':3,'ARROZ':4,'ACEITE':0,'ATUN':2}
BEN={1:120,2:140,3:110}
#Definición del modelo
modelo = LpProblem("Conformación_de_Kits_V2",LpMaximize)

modelo+=lpSum(BEN[j]*k[(j)] for j in kits)


#Restricciones
#for i in alimento:
    #modelo+=lpSum(C[j][i]*k[(j)] for j in kits)<=INV[i]

for i in alimento:
    modelo+=lpSum(C[j][i]*k[(j)] for j in kits)+HINV[i]==INV[i]

for i in alimento:
    for j in kits:
        modelo+=X[(i,j)]==C[j][i]*k[(j)]
    
        
modelo.solve()
print(LpStatus[modelo.solve()])
if LpStatus[modelo.solve()]=="Optimal":
    print(value(modelo.objective))
    for v in modelo.variables():
            print(v.name, "=", v.varValue)
            
    else:
        print(LpStatus[modelo.solve()])
```

```python
#Primera parte con holgura para kits suplementarios
#DEFINICIÓN DE ÍNDICES
#Indice i
alimento=['AZUCAR','ARROZ','ACEITE','ATUN']
#Indice j
kits=[1,2,3]
#Indice k
suplemento=['PANELA','LECHE']
#Indice l
kits_sec=[1]

#VARIABLES DE DECISIÓN
#Cantidad de alimento principal i asignado al kit principal j
XP = LpVariable.dicts("Cantidad_Principal",[(i,j)for i in alimento for j in kits],lowBound=0,cat='Integer')
#Cantidad de kits principales del tipo j a conformar 
YP = LpVariable.dicts("Kit_Principales",[(j)for j in kits],lowBound=0,cat='Integer')
#Cantidad de alimento suplementario k asignado al kit secundario l
XS=LpVariable.dicts("Cantidad_Suplemento",[(k,l)for k in suplemento for l in kits_sec],lowBound=0,cat='Integer')
#Cantidad de kits secundarios del tipo l a conformar 
YS = LpVariable.dicts("Kit_Secundarios",[(l)for l in kits_sec],lowBound=0,cat='Integer')
HINV=LpVariable.dicts("Holgura_Inventario",[(i)for i in alimento],lowBound=0,cat='Integer')


#PARÁMETROS
#Cantidad de alimento i empleado en conformar el kit tipo j
C={1:{'AZUCAR':1,'ARROZ':2,'ACEITE':0,'ATUN':1},
  2:{'AZUCAR':2,'ARROZ':0,'ACEITE':1,'ATUN':1 },
  3:{'AZUCAR':1,'ARROZ':1,'ACEITE':0,'ATUN':1 }}

#Cantidad de alimento sumplementario k empleado en conformar el kit suplementario tipo l
CS={1:{'PANELA':1,'LECHE':1}}

#Cantidad de alimento principal i empleado en conformar el kit suplementario tipo l
CSP={1:{'AZUCAR':0,'ARROZ':1,'ACEITE':0,'ATUN':0}}

INV={'AZUCAR':3,'ARROZ':4,'ACEITE':0,'ATUN':2}
INVS={'PANELA':40,'LECHE':60}
BEN={1:120,2:140,3:110}
BENS={1:80}
#Definición del modelo
modelo = LpProblem("Conformación_de_Kits_V2",LpMaximize)

modelo+=lpSum(BEN[j]*YP[(j)] for j in kits) +lpSum(BENS[l]*YS[(l)] for l in kits_sec)


#Restricciones
for i in alimento:
    modelo+=lpSum(C[j][i]*YP[(j)] for j in kits)+HINV[i]==INV[i]

for i in alimento:
    for j in kits:
        modelo+=XP[(i,j)]==C[j][i]*YP[(j)]

for i in alimento:
    modelo+=lpSum(CSP[l][i]*YS[(l)] for l in kits_sec)<=HINV[i]

for k in suplemento:
    modelo+=lpSum(CS[l][k]*YS[(l)] for l in kits_sec)<=INVS[k]

modelo.solve()
print(LpStatus[modelo.solve()])
if LpStatus[modelo.solve()]=="Optimal":
    print(value(modelo.objective))
    for v in modelo.variables():
            print(v.name, "=", v.varValue)
            
    else:
        print(LpStatus[modelo.solve()])
```
